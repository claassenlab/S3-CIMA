```python
import scanpy as sc
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
import matplotlib as mpl
import matplotlib.cm as cm

from sklearn.metrics.pairwise import pairwise_distances
import scipy.stats as ss

import anndata as ad
```

```python
sc.set_figure_params(scanpy=True, dpi=150, dpi_save=300, frameon=True, vector_friendly=True, fontsize=10, figsize=(4,4),
                     color_map=None, format='pdf', facecolor=None, transparent=True, ipython_format='png2x')

sc.settings.figdir = "scRNA-seq/Redone22/Manuscript/Zheng"

mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
```

## Caveats:

- We lost 35 genes due to missing geneSymbols

```python
df = pd.read_csv("Zheng_Study/GSE98638_HCC.TCell.S4070.norm.centered.txt", delimiter="\t")
df = df.dropna()
colnames_df = df["geneSymbol"].copy()
df = df.drop(["geneID", "geneSymbol"], axis=1)
df = df.T
df.columns = colnames_df
df.head()
```

```python
df_raw = pd.read_csv("Zheng_Study/GSE98638_HCC.TCell.S5063.count.txt", delimiter="\t")
```

```python
df_raw = df_raw.dropna()
ddf_rawf = df.dropna()
colnames_df_raw = df_raw["symbol"].copy()
df_raw = df_raw.drop(["geneID", "symbol"], axis=1)
df_raw = df_raw.T
df_raw.columns = colnames_df_raw
df_raw.head()
```

```python
meta_full = pd.read_csv("Zheng_Study/meta.txt", delimiter=",")
```

```python
meta = meta_full.loc[meta_full["UniqueCell_ID"].isin(df.index.to_list())].copy()
meta_raw = meta_full.loc[meta_full["UniqueCell_ID"].isin(df_raw.index.to_list())].copy()
```

```python
adata_raw = ad.AnnData(df_raw.values, dtype=np.float32)
adata_raw.obs_names = meta_raw["UniqueCell_ID"]
adata_raw.var_names = df_raw.columns
adata_raw.obs["Celltype"] = pd.Categorical(meta_raw["majorCluster"])
adata_raw.obs["Patient"] = pd.Categorical(meta_raw["Patient"])
```

```python
adata = ad.AnnData(df.values, dtype=np.float32)
```

```python
adata.obs_names = meta["UniqueCell_ID"]
adata.var_names = df.columns
adata.obs["Celltype"] = pd.Categorical(meta["majorCluster"])
adata.obs["Patient"] = pd.Categorical(meta["Patient"])
```

```python
adata = adata[adata.obs.Celltype == "C03_CD8-SLC4A10"].copy()
```

```python
shared_genes = adata.var_names.intersection(adata_raw.var_names)
adata_raw = adata_raw[adata.obs_names, shared_genes].copy()
adata = adata[:, shared_genes].copy()
```

```python
sc.pp.normalize_total(adata_raw, target_sum=1e4)
sc.pp.log1p(adata_raw)
```

```python
# Make sure the ordering is identical
(adata.obs_names == adata_raw.obs_names).all() & (adata.var_names == adata_raw.var_names).all()
```

```python
adata.layers["RAW"] = adata_raw.X.copy()
```

```python
adata.var["mt"] = adata.var_names.str.startswith("MT-")
adata.var["ribo"] = adata.var_names.str.startswith(("RPS", "RPL"))
adata.var["hb"] = adata.var_names.str.contains(("^HB[^(P)]"))
```

```python
sc.pp.calculate_qc_metrics(
        adata, qc_vars=["mt", "ribo", "hb"], percent_top=None, log1p=False, inplace=True)
```

```python
malat1 = adata.var_names.str.startswith("MALAT1")

mito_genes = adata.var_names.str.startswith("MT-")
ribo_genes = adata.var_names.str.contains(("^RP[SL]"))
hb_genes = adata.var_names.str.contains("^HB[^(P)]")


remove = np.add(mito_genes, malat1)
remove = np.add(remove, ribo_genes)
remove = np.add(remove, hb_genes)

keep = np.invert(remove)

adata = adata[:, keep]
```

```python
sc.pp.highly_variable_genes(adata)
```

```python
sc.pp.scale(adata)
```

```python
sc.pp.pca(adata, n_comps=50, use_highly_variable=True, random_state=0)
```

```python
sc.pp.neighbors(adata, n_neighbors=30, n_pcs=20, random_state=0)
```

```python
sc.tl.umap(adata, random_state=0)
```

```python
res_lst = [0.6, 0.7, 0.8, 0.9]
for resolution in res_lst:
    sc.tl.leiden(adata, key_added=f"leiden_r{resolution}", resolution=resolution, random_state=0)
```

```python
with pd.ExcelWriter(f"scRNA-seq/Redone22/Manuscript/Zheng/cluster_sizes.xlsx") as writer:
    for resolution in res_lst:
        count_df = adata.obs[f"leiden_r{resolution}"].value_counts().to_frame()
        count_df.to_excel(writer, sheet_name=f"r{resolution}")
```

```python
## Load the profile of the MAIT dysfunctional cluster

df = pd.read_excel(io="scRNA-seq/Redone22/MAIT_Bootstrap/3/DGE/leiden_r0.4_bootstrap.xlsx", sheet_name="c4")
df.drop(["Unnamed: 0", "group"], axis=1, inplace=True)
```

```python
shared_mait_genes = adata.var_names.intersection(df.names)
df = df[df.names.isin(shared_mait_genes)]
```

```python
adata_dys = sc.read("scRNA-seq/Redone22/MAIT_Bootstrap/3/Data/MAIT_BOOTSTRAPPED_3.h5ad")
```

```python
res = pd.DataFrame(columns=adata_dys.var_names, index=adata_dys.obs["leiden_r0.4"].cat.categories)                                                                                                 

for clust in adata_dys.obs["leiden_r0.4"].cat.categories: 
    res.loc[clust] = adata_dys[adata_dys.obs["leiden_r0.4"].isin([clust]),:].X.mean(0)
    
for col in res.columns:
    res[col] = res[col].astype(np.float32)
```

```python
number_top_genes = 25
comp_genes = df.sort_values(by="scores", key=abs, ascending=False).head(number_top_genes).names.to_list()
```

```python
res = res[comp_genes].copy()
```

```python
adata = adata[:, comp_genes].copy()
```

```python
for zheng_clustering_res in res_lst:
    res_zheng = pd.DataFrame(columns=adata.var_names, index=adata.obs[f"leiden_r{zheng_clustering_res}"].cat.categories)                                                                                                 

    for clust in adata.obs[f"leiden_r{zheng_clustering_res}"].cat.categories: 
        res_zheng.loc[clust] = adata[adata.obs[f"leiden_r{zheng_clustering_res}"].isin([clust]),:].X.mean(0)
    
    for col in res_zheng.columns:
        res_zheng[col] = res_zheng[col].astype(np.float32)

    res_zheng = res_zheng[comp_genes].copy()
    
    res.index = [f"c{idx}" for idx,_ in enumerate(res.index)]
    res_zheng.index = [f"z{idx}" for idx,_ in enumerate(res_zheng.index)]
    
    full_df = pd.concat([res, res_zheng])
    corr_df = full_df.T.corr(method="pearson").iloc[:len(res.index),len(res.index):]
    
    sns.heatmap(data=corr_df, annot=True, center=0, cmap="RdBu_r", linewidths=0.2, linecolor="black")
    plt.title(f"Pearson correlation between clusters. Resolution={zheng_clustering_res}, {number_top_genes} genes")
    plt.savefig(f"scRNA-seq/Redone22/Manuscript/Zheng/{number_top_genes}_genes_res-{zheng_clustering_res}.pdf", 
                dpi=400, transparent=True, bbox_inches="tight")
    plt.close()
```
